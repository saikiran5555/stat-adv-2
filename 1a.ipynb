Probability Density Function (PDF)
The Probability Density Function (PDF), on the other hand, is used for continuous random variables. These are variables that can take on an infinite number of values within a given range. The PDF describes the likelihood of the random variable falling within a particular range of values, as opposed to taking on any one specific value.

Properties of PDF:

The area under the PDF (and above the horizontal axis) across all possible values of the random variable is equal to 1.
The PDF itself is not a probability, but an infinitesimally small area under the curve can be interpreted as a probability.

Key Differences
Type of Variable: PMF is used for discrete random variables, while PDF is used for continuous random variables.
Interpretation: In PMF, the value of the function gives the probability of the random variable being exactly a certain value. In PDF, the value of the function is not a probability but rather a density; the probability is given by the area under the PDF curve within a range of values.
Sum vs. Integral: The total of all probabilities in a PMF (sum over all possible values) is 1, while the total area under a PDF curve (integral over all possible values) is 1.
