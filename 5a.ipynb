Focus:

Bernoulli Distribution: Focuses on the outcome of a single event (success or failure).
Binomial Distribution: Focuses on the number of successes in a given number of trials.
Application:

Bernoulli Distribution: Used when there's only one event (e.g., a single coin toss).
Binomial Distribution: Used when the same event is repeated multiple times (e.g., multiple coin tosses).
